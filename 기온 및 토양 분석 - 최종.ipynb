#### 봄 밀의 온도 조건
- 1-3℃에 싹트며, 제일 좋은 싹 트기 온도는 15-20℃ 이다.
- 생육 기간 15℃ 이상의 평균 기온만 보장되면 어디서나 재배할 수 있다.

#### 봄밀의 토양 조건
- 봄밀이 자랄 때에는 토양 물기가 70-80%가 적정 수치이다.
- 우리나라 북부고지대는 밀이 자라는 기간에 580-680mm의 비가 내리므로 봄밀의 물기 요구를 충분히 보장해주고 있다.
- 토양의 산도는 pH 6-7이 좋다.

#### 가을밀의 온도조건
- 가을밀은 봄밀보다 높은 온열조건을 요구한다.
- 밀은 3-4℃에서 싹트기 시작하며 30℃이상의 높은 온도가 계속되면 생육에 좋지 않은 영향을 받는다.
- 가을 밀은 초겨울에 추위견딜성이 비교적 약하여 -9℃에서도 얼어죽을 수 있으나 점차 추위 견딜성이 강해져서 겨울에는 -15~-17℃에서도 견딜 수 있다.
- 가을 밀보리는 1월 평균 최저기온이 -13℃ 이상 되는 지역까지 배치할 수 있다.
- 추위 견딜성이 강한 밀은 1월 평균 최저기온이 -15℃이상 되는 지역에도 배치할 수 있다.

#### 밀의 해비침 및 토양 조건
- 밀은 장일성식물이며 빛을 많이 요구하는 시기는 꽃이 펴서 이삭이 생기는 시기까지이다.
- 밀밭은 땅이 걸고 물이 잘 빠지며 갈이층이 깊고 유기질과 인산, 알칼리 함량이 많은 모래메흙, 메흙, 질메흙 땅에 배치하는 것이 좋다.
- 토양의 반응은 중성 및 약산성(pH6.0 - 7.5)이 좋으며 산성 토양에서는 밀이 잘 자라지 못한다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

# knn 모델 기법

```python
# knn 알고리즘 사용함
# 밑에 나온 결과는 2020년 기준으로 학습한 데이터로 검증한 정확도이다,

from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()
```

### 교차검증 - KNN, Decision Tree 검증, 잘 학습했는지
- KNN 알고리즘을 사용하고 단편적으로 정확도를 판단했는데, 한 번 더 데이터를 섞어서 재구성한다음 정확도를 출력함
- DecisionTree 알고리즘 사용하고 단편적으로 정확도를 판단했는데, 한 번 더 데이터를 섞어서 재구성한다음 정확도를 출력함

```python
# 교차 검증 기능을 이용하기 위한 준비물들을 import 한다.

from sklearn.model_selection import cross_val_score
```

```python
# knn 알고리즘 검증 시작
# 결과 값은 array 값이라 출력된 것이 각각 검증한 정확도이다.
# 저거 검증 횟수는 cv = n 값을 수정하면 더 많이 시행할 수도, 적게 시행할 수 있다.

knn_result = cross_val_score(knn_model,
                            X_train,
                            y_train,
                            cv = 5)

knn_result
```

```python
# Decision Tree 검증 시작
# 결과 값은 array 값이라 출력된 것이 각각 검증한 정확도이다.
# 저거 검증 횟수는 cv = n 값을 수정하면 더 많이 시행할 수도, 적게 시행할 수 있다.


tree_result = cross_val_score(tree_model,
                             X_train,
                             y_train,
                             cv = 5)

tree_result
```

```python
# 위에 나온 결과 값들의 평균을 낸다

tree_result.mean()
```

### RandomForest 기법

```python
# Random Forest 기법에 사용할 준비물 import
# 그리고 학습 시작

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
                                 random_state = 320,
                                 n_estimators = 1000, # 생성될 트리 개수(단일 모델의 개수)
                                 max_depth = 10, # 트리의 깊이
                                 max_features=0.5, # 사용할 컬럼 개수의 비율
                                 )

rf_model.fit(X_train, y_train)
```

```python
# 랜던 포레스트 학습 결과 나온다.
# 학습 완료한 정확도가 5개 출력됨

rf_result1 = cross_val_score(rf_model,
               X_train,
               y_train,
               cv = 5)

rf_result1
```

```python
# 5개 정확도의 평균을 낸다.

rf_result1.mean()
```

```python
# 교차검증을 진행해 이게 잘 학습했는지 한 번 더 확인한다.
# 확인한 정확도가 출력된다.

cross_val_score(rf_model, X_train, y_train, cv = 5).mean()
```
